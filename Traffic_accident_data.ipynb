## Checking for null/missing values in the dataset

```python
df.isna().sum()
```

## Dropping the columns with huge number of null values

```python
df.drop(columns=['End_Lat','End_Lng','Wind_Chill(F)','Precipitation(in)'],axis=1,inplace=True)
```

```python
#storing categorical column names to a new variable
categorical=[i for i in df.columns if df[i].dtype=='O']
#for categorical values we can replace the null values with the Mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.drop(columns=['Wind_Speed(mph)', 'Visibility(mi)', 'Pressure(in)', 'Humidity(%)', 'Temperature(F)'], axis=1, inplace=True)
```

# EDA

## Accident Severity Distribution

```python
# Count the number of occurrences of each severity level
severity_counts = df['Severity'].value_counts().sort_index()

# Create labels and sizes for the pie chart
labels = ["Least impact", "Moderate impact", "High impact","Significant impact"]
sizes = severity_counts.values
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']  # Optional: custom colors

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=60, colors=colors)
plt.title('Accident Severity Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

## City with most Number of accidents

```python
city_acc = df['City'].value_counts().sort_values(ascending = False).reset_index()
```

```python
city_acc
```

## Top 10 cities with most number of accidents

```python
plt.figure(figsize=(15,6))
sns.barplot(x='City',y='count',data= city_acc.head(10),palette='rocket')
plt.title("Top 10 cities with most number of accidents")
plt.ylabel("No of accidents")
plt.show()
```

## Bottom 10 cities with least number of accidents

```python
plt.figure(figsize=(15,6))
sns.barplot(x='City',y='count',data= city_acc.tail(10),palette='rocket')
plt.title("Bottom 10 cities with least number of accidents")
plt.ylabel("No of accidents")
plt.xticks(rotation=45)
plt.show()
```

## State with most and least Number of accidents

```python
state_acc = df['State'].value_counts().sort_values(ascending = False).reset_index()
state_acc
```

```python
## States with most number of accidents
plt.figure(figsize=(15,6))
sns.barplot(x='State',y='count',data= state_acc.head(5),palette='rocket')
plt.title("Top 5 States with most number of accidents")
plt.ylabel("No of accidents")
plt.show()

## States with the least number of accidents
plt.figure(figsize=(15,6))
sns.barplot(x='State',y='count',data= state_acc.tail(5),palette='rocket')
plt.title("Bottom 5 States with least number of accidents")
plt.ylabel("No of accidents")
plt.show()
```

`Observations:`
- Miami is the city with most number of accidents
- starjunction, stomsburg are among the cities with least number of accidents
- California is the state with highest cases of accident
- South Dakota(SD) is the state with least cases of accident

```python
# Convert the 'Start_Time' column to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')

# Extract the year from the 'Start_Time' column and store it in a new column called 'Year'
df['Year'] = df['Start_Time'].dt.year

# Display the first few rows to verify
print(df[['Start_Time', 'Year']].head())
```

`Observation:`

# Conclusion:

This traffic data analysis project involved a detailed review of accident records, including data cleaning, exploratory data analysis (EDA), and visualization. The main findings from the analysis are outlined below:

1. **City-wise Accident Distribution**:
   - **Miami** had the highest number of accidents, likely due to its large population and high traffic density.
   - Smaller cities like **Star Junction** and **Stromsburg** reported the fewest accidents, reflecting their lower traffic and population levels.

2. **State-wise Accident Distribution**:
   - **California** recorded the most accidents, consistent with its large population and extensive road network.
   - **South Dakota (SD)** had the fewest accidents, likely due to its lower population density and reduced traffic volume.

3. **Yearly Accident Trends**:
   - The year **2021** saw the highest number of accidents, suggesting either an increase in traffic or changes in accident reporting during that period.

4. **Accident Severity**:
   - Accidents were categorized into four severity levels: 1, 2, 3, and 4.
   - Most accidents were classified as **severity level 2**, indicating that while accidents were common, they were generally not highly severe.

5. **Weather Conditions During Accidents**:
   - A majority of accidents occurred in **fair weather**, indicating that other factors like driver behavior or traffic congestion were likely more influential in causing accidents than poor weather conditions.

These findings offer valuable insights into accident patterns and locations, providing crucial information for shaping policy, enhancing urban planning, and implementing safety measures to reduce both the occurrence and impact of traffic accidents.
